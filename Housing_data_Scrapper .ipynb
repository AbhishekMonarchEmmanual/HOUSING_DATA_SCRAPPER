```python
page = BeautifulSoup(page.content, "html.parser")
```

```python
lists = page.find_all("section", {"class":"listing-search-item listing-search-item--list listing-search-item--for-rent"})
```

```python
with open("housing_data.csv", "w", encoding = "utf-8", newline= '') as f:
    thewriter= writer(f)
    header = ["TITLE", "LOCATIONS", "INTERIORS","NUMBER OF ROOMS", "APT_SIZE","PRICE"]

    thewriter.writerow(header)



    for i in range(len(lists)):
        title = lists[i].find_all("a", {"class": "listing-search-item__link listing-search-item__link--title"})[0].text.strip()
        locations = lists[i].find("div", {"class":"listing-search-item__sub-title"}).text.strip()
        interior = lists[i].find("li", {"class": "illustrated-features__item illustrated-features__item--interior"}).text.strip()
        no_rooms= lists[i].find("li", {"class": "illustrated-features__item illustrated-features__item--number-of-rooms"}).text.strip()
        apt_size= lists[i].find("li", {"class": "illustrated-features__item illustrated-features__item--surface-area"}).text.strip()
        price = lists[i].find("div", {"class": "listing-search-item__price"}).text.strip()
        l = [price, title,locations, interior, no_rooms,apt_size]
        thewriter.writerow(l)
        
    
```
